```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
from skimage import io, transform
import numpy as np

# === Load Image using Pillow ===
img = Image.open("images.jpeg")

# === Pillow Operations ===
img_resized = img.resize((200, 200))
img_rotated = img.rotate(45)
img_flipped = ImageOps.mirror(img)
img_cropped = img.crop((50, 50, 300, 300))  # (left, top, right, bottom)

# === Display Pillow Results ===
fig, axs = plt.subplots(1, 4, figsize=(15, 4))
axs[0].imshow(img_resized)
axs[0].set_title("Resized (Pillow)")
axs[1].imshow(img_rotated)
axs[1].set_title("Rotated (Pillow)")
axs[2].imshow(img_flipped)
axs[2].set_title("Flipped (Pillow)")
axs[3].imshow(img_cropped)
axs[3].set_title("Cropped (Pillow)")
for ax in axs:
    ax.axis('off')
plt.tight_layout()
plt.show()

```

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
from skimage import io, transform
import numpy as np

# === Load Image using scikit-image ===
img_sk = io.imread("images.jpeg")

# === scikit-image Transformations ===
img_sk_resized = transform.resize(img_sk, (200, 200))
img_sk_rotated = transform.rotate(img_sk, 45, resize=True)  # rotates and expands canvas

# === Display scikit-image Results ===
fig2, axs2 = plt.subplots(1, 2, figsize=(10, 4))
axs2[0].imshow(img_sk_resized)
axs2[0].set_title("Resized (scikit-image)")
axs2[1].imshow(img_sk_rotated)
axs2[1].set_title("Rotated (scikit-image)")
for ax in axs2:
    ax.axis('off')
plt.tight_layout()
plt.show()
```
