```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# read and display the color image
img = cv2.imread("2637588.jpg")
# Display
cv2_imshow(img)
```

```python
# Plot histograms for R, G, B channels
color = ('r','g','b')
for i,col in enumerate(color):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color = col)
    plt.xlim([0,256])
plt.title('Histogram of Original Image')
plt.show()
```

```python
# Split channels
r, g, b = cv2.split(img)
```

```python
# Equalize each channel separately
r_eq = cv2.equalizeHist(r)
g_eq = cv2.equalizeHist(g)
b_eq = cv2.equalizeHist(b)
```

```python
# Merge equalized channels
equalized_img = cv2.merge((r_eq, g_eq, b_eq))
```

```python
# Display result
plt.subplot(1,2,1)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
equalized_img_rgb = cv2.cvtColor(equalized_img, cv2.COLOR_BGR2RGB)
plt.imshow(equalized_img_rgb)
plt.title('Histogram Equalized Image')
plt.axis('off')
plt.show()
```

```python
# Plot histograms for Equalised R, G, B channels
color = ('r','g','b')
for i,col in enumerate(color):
    hist = cv2.calcHist([equalized_img_rgb],[i],None,[256],[0,256])
    plt.plot(hist,color = col)
    plt.xlim([0,256])
plt.title('Histogram Equalised')
plt.show()
```

```python
# Plot histograms for Equalised R, G, B channels
color = ('r','g','b')
for i,col in enumerate(color):
    hist = cv2.calcHist([equalized_img_rgb],[i],None,[256],[0,256])
    plt.plot(hist,color = col)
    plt.xlim([0,256])
plt.title('Histogram Equalised')
plt.show()
```

```python
# Create CLAHE object
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))

# Apply CLAHE on each channel
r_clahe = clahe.apply(r)
g_clahe = clahe.apply(g)
b_clahe = clahe.apply(b)

# Merge CLAHE enhanced channels
clahe_img = cv2.merge((r_clahe, g_clahe, b_clahe))
```

```python
# Display result
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
clahe_img_rgb = cv2.cvtColor(clahe_img, cv2.COLOR_BGR2RGB)
plt.imshow(clahe_img_rgb)
plt.title('CLAHE Enhanced Image')
plt.axis('off')
plt.show()
```
